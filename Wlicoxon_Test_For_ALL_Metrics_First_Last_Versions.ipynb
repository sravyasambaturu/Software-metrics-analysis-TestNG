```python
from google.colab import files
import pandas as pd
import io

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))

  # Set header=0 to use the first row as column names
  df = pd.read_excel(io.BytesIO(uploaded[fn]), sheet_name='MaxCyclomatic', header=0)

df
```

```python
from scipy.stats import wilcoxon
import itertools
import pandas as pd # Added import for pandas


# Assuming 'df' is your DataFrame
# Convert the column names to strings if they are not already
df.columns = df.columns.astype(str)

# Your existing code
columns_to_test = ['7.11', '7.5', '6.13.1', '6.0.1', '5.13']

results = []
for col1, col2 in itertools.combinations(columns_to_test, 2):
    statistic, p_value = wilcoxon(df[col1], df[col2])
    if p_value < 0.05:
        significance = "Significant difference detected."
    else:
        significance = "No significant difference detected."
    results.append([f"{col1} vs {col2}", statistic, p_value, significance])


df_results = pd.DataFrame(results, columns=["Version Combo", "T-value", "P-value", "Significance"])
df_results
```
