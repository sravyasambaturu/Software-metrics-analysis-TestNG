```python
from google.colab import files
import pandas as pd
import io


uploaded = files.upload()


for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))


  df = pd.read_excel(io.BytesIO(uploaded[fn]), sheet_name='Wilcoxon_LOC_All _Versions')


df
```

```python
from scipy.stats import wilcoxon
import itertools

columns_to_test = ['Version_7.11', 'Version_7.5', 'Version_6.13.1', 'Version_6.0.1', 'Version_5.13']

# for col1, col2 in itertools.combinations(columns_to_test, 2):
#     statistic, p_value = wilcoxon(df[col1], df[col2])
#     print(f"Wilcoxon test for {col1} vs {col2}:")
#     print(f"  T-value: {statistic}")
#     print(f"  P-value: {p_value}")
#     if p_value < 0.05:
#         print("  Significant difference detected.")
#     else:
#         print("  No significant difference detected.")
#     print("-" * 20)

results = []
for col1, col2 in itertools.combinations(columns_to_test, 2):
    statistic, p_value = wilcoxon(df[col1], df[col2])
    if p_value < 0.05:
        significance = "Significant difference detected."
    else:
        significance = "No significant difference detected."
    results.append([f"{col1} vs {col2}", statistic, p_value, significance])


df_results = pd.DataFrame(results, columns=["Version Combo", "T-value", "P-value", "Significance"])
df_results
```
